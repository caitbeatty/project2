```python
import csv, json
li = []
with open('merged_df.csv', newline='') as csvfile:
    reader = csv.reader(csvfile, delimiter=',')
    for Household_Income, ID_Year, geoid, latitude, longitude in reader:
        li.append({
            "Household_Income": Household_Income,
           "ID_Year":ID_Year,
            "geoid" :geoid,
           "latitude": latitude,
           "longitude": longitude,
           "geo": {
                "__type": "GeoPoint",
                "latitude": latitude,
                "longitude": longitude,
            }
        })
with open("merged_df.json", "w") as f:
    json.dump(li, f)
```
